## COVID-19 백신 종류별 접종 인구

```python
days = [1, 2, 3] # 1일, 2일, 3일
az = [2, 4, 8] # 1일부터 3일까지 백신 접종 인구(만)
pfizer = [5, 1, 3]
moderna = [1, 2, 5]
```

```python
plt.plot(days, az, label="AZ", color="r",
        marker='o', ls='--')
plt.plot(days, pfizer, label="PFIZER", color ="g",
        marker='o')
plt.plot(days, moderna, label="MODRENA", color="b",
        marker='o')
plt.legend(ncol=3)
plt.show()
```
