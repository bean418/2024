```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
df=pd.read_csv('boston.csv')
df['price']=df['medv']
df.head()
```

```python
input=df.iloc[:,:-1]
output=df['price']
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_output, test_output=train_test_split(input,output)
```

```python
from xgboost import XGBRegressor, plot_importance
```

```python
XGB=XGBRegressor(max_depth=3, reg_lambda=0.1) #L2 #reg_alpha=5 (L1)
```

```python
XGB.fit(train_input,train_output)
```

```python
np.corrcoef(XGB.predict(test_input),test_output)
```

```python
plot_importance(XGB,max_num_features=10).set_yticklabels(df.columns)
```
