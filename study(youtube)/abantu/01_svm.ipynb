```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
cancer = datasets.load_breast_cancer()
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['degree'] = cancer.target
```

```python
df.head()
```

```python
np.unique(df['degree'])
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
cancer_scaled = SS.fit_transform(cancer.data)
cancer_scaled
```

```python
input = cancer_scaled
output = df['degree']
```

```python
from sklearn.model_selection import train_test_split
train_in, test_in, train_out, test_out = train_test_split(input, output,
                                                         random_state=0)
```

```python
from sklearn import svm
svc = svm.SVC(kernel='linear', C=0.1)
svc2 = svm.SVC(kernel='poly', C=0.1)
svc3 = svm.SVC(kernel='rbf', C=0.1)
```

```python
svc.fit(train_in,train_out)
svc2.fit(train_in,train_out)
svc3.fit(train_in,train_out)
```

```python
svc.score(train_in,train_out),svc.score(test_in,test_out)
```

```python
svc2.score(train_in,train_out),svc2.score(test_in,test_out)

# under 85% -> underfitting : 과소적합
# 유연함이 있다?
```

```python
svc3.score(train_in,train_out),svc3.score(test_in,test_out)
```

```python
# 예외 허용 X
svc = svm.SVC(kernel='linear', C=100)
svc2 = svm.SVC(kernel='poly', C=100)
svc3 = svm.SVC(kernel='rbf', C=100)
```

```python
svc.fit(train_in,train_out)
svc2.fit(train_in,train_out)
svc3.fit(train_in,train_out)
```

```python
svc.score(train_in,train_out),svc.score(test_in,test_out)

# train 정확도는 매우 높음, test 정확도는 떨어짐
# OverFitting : 과적합
# 족보를 매우 열심히 보았지만, 족보 이외의 문제는 풀지 못 하는 상황
# 
```
